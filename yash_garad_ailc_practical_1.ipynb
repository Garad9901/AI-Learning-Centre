## Generate a triangle of stars

### Subtask:
Write code to print a triangle pattern using star characters and loops.


**Reasoning**:
The subtask requires generating a triangle (pyramid) pattern using star characters and loops, following the provided step-by-step instructions. I will write a Python code block to implement this logic.



````python
num_rows = 5

for i in range(1, num_rows + 1):
    # Print leading spaces
    print(" " * (num_rows - i), end="")
    # Print stars
    print("*" * (2 * i - 1))
````

````python
n = 5   # number of rows

for i in range(n):
    # print spaces
    print("  " * i, end="")

    # print alphabets
    for j in range(n - i):
        print(chr(65 + j), end="  ")

    print()
````

````python
# Code for the 'c' pattern using a loop
num_rows_c_pattern = 5
c_char = 'c'

for i in range(num_rows_c_pattern):
    if i == 0 or i == num_rows_c_pattern - 1:
        # Top and bottom line: 8 'c's separated by spaces
        print((c_char + " ") * 7 + c_char)
    else:
        # Middle lines: single 'c'
        print(c_char)
````

````python
# Code for the numbered diamond pattern using a loop
num_rows_diamond_pattern = 5

for i in range(1, num_rows_diamond_pattern + 1):
    if i == 1:
        print(f"{i}{' ' * 17}{i}")
    elif i == 2:
        print(f"{' ' * 4}{i}{' ' * 8}{i}")
    elif i == 3:
        print(f"{' ' * 8}{i}")
    elif i == 4:
        print(f"{' ' * 3}{i}{' ' * 9}{i}")
    elif i == 5:
        print(f"{i}{' ' * 18}{i}")
````
